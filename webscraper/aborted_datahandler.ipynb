```python
web = [i.replace('</a>','') for i in web ]
web = [i.replace('href=','') for i in web ]
```

```python
websites = [i.split('>')[0] for i in web ]
names = [i.split('>')[1] for i in web ]
```

```python
websites =['http://www.iachina.cn'+i.replace('"','') for i in websites]
```

```python
names_prefix = []
for i in names:
    names_year_month = re.findall(r'\d{4}年\d{1,2}-\d{1,2}月|\d{4}年\d{1,2}月|\d{4}年', i)
    names_prefix.append(names_year_month[0])
```

```python
df = pd.DataFrame(columns = ['Class','TotalEarning','date_time'])
for i in range(len(websites)):
    url = websites[0]
    r = requests.get(url)
    soup = BeautifulSoup(r.content,'html.parser')
    table = soup.find('table',border = '1')
    for row in table.tbody.find_all('tr')[2:]:
        columns = row.find_all('td')
        if (columns[0].text.strip() != ''):
            class_ = columns[0].text.strip()
            earning = columns[1].text.strip()
            date = names_prefix[i]
        df = df.append({'Class': class_, 'TotalEarning':earning,'date_time':date},ignore_index=True)
```
